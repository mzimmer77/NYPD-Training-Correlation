```python
import pandas as pd
from datetime import datetime

# Define the start date (January 1, 1980)
start_date = datetime(1980, 1, 1)

# Function to calculate days since 1980
def days_since_1980(date):
    return (date - start_date).days

# Read the CSV files
training_data = pd.read_csv('training_clean.csv', on_bad_lines='skip')
complaint_data = pd.read_csv('Civilian_Complaint_Review_Board__Allegations_Against_Police_Officers_20240706.csv', on_bad_lines='skip')
officer_data = pd.read_csv('officer_pages.csv', on_bad_lines='skip')

# Convert the taxid columns to strings to ensure consistent merging
training_data['taxid'] = training_data['taxid'].astype(str)
complaint_data['Tax ID'] = complaint_data['Tax ID'].astype(str)
officer_data['officers'] = officer_data['officers'].astype(str)

# Filter training data for entries that include "investigative" in the training name
investigative_trainings = training_data[training_data['name'].str.contains('investigative', case=False, na=False)]

# Print to inspect the filtered trainings
print(investigative_trainings.head())

# Merge filtered training data with officer data on taxid (officers)
merged_data = pd.merge(investigative_trainings, officer_data, left_on='taxid', right_on='officers', how='inner')

# Print to inspect the merged data
print(merged_data.head())

# Select only the first 5 officers
first_five_officers = merged_data.head(5)

# Print to inspect the first 5 officers
print(first_five_officers)

# Convert relevant date columns to datetime and calculate days since 1980
complaint_data['As Of Date'] = pd.to_datetime(complaint_data['As Of Date'])
first_five_officers['training_date'] = pd.to_datetime(first_five_officers['date'])

# Calculate days since 1980 for both training and complaint dates
first_five_officers['days_since_1980'] = first_five_officers['training_date'].apply(days_since_1980)
complaint_data['days_since_1980'] = complaint_data['As Of Date'].apply(days_since_1980)

# Merge first_five_officers with complaint data on taxid
final_merged = pd.merge(first_five_officers, complaint_data, left_on='taxid', right_on='Tax ID', how='inner')

# Calculate the days between training and complaints
final_merged['days_after_training'] = final_merged['days_since_1980_y'] - final_merged['days_since_1980_x']

# Filter complaints that happened within 182 days after training
final_filtered = final_merged[(final_merged['days_after_training'] >= 0) & (final_merged['days_after_training'] <= 182)]

# Print the resulting DataFrame
print(final_filtered[['taxid', 'name', 'training_date', 'As Of Date', 'days_after_training', 'complaint', 'allegation', 'substantiated']])

# Save the result to a CSV file
final_filtered.to_csv('complaints_and_times.csv', index=False)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Load the CSV files
training_data = pd.read_csv('training_clean.csv', on_bad_lines='skip')
complaints_data = pd.read_csv('combined_allegations_complaints.csv', on_bad_lines='skip')

# Drop the unwanted timestamp column
training_data = training_data.drop(columns=['timestamp..2024.07.07'])

# Convert 'date' column in training_data to datetime
training_data['date'] = pd.to_datetime(training_data['date'])

# Filter training data for May 2017
start_date = datetime(2017, 5, 1)
end_date = datetime(2017, 5, 31)
filtered_training_data = training_data[(training_data['date'] >= start_date) & (training_data['date'] <= end_date)]

# Convert 'Incident Date' column in complaints_data to datetime
complaints_data['Incident Date'] = pd.to_datetime(complaints_data['Incident Date'], format='%Y/%m/%d')

# Define the date range for incidents (6 months after May 2017)
incident_start_date = end_date + timedelta(days=1)
incident_end_date = incident_start_date + timedelta(days=182)  # 6 months

# Filter complaints data for incidents within 6 months after May 2017
filtered_complaints_data = complaints_data[(complaints_data['Incident Date'] >= incident_start_date) & (complaints_data['Incident Date'] <= incident_end_date)]

# Merge the filtered training data with the filtered complaints data on 'Tax ID'
combined_data = pd.merge(filtered_training_data, filtered_complaints_data, left_on='taxid', right_on='Tax ID', how='inner')

# Reorder columns to have 'Tax ID' first
combined_data = combined_data[['Tax ID', 'date', 'name', 'clt', 'srg', 'fts', 'cjb', 'investigative', 'patrol', 'camera', 'terrorism', 'carstop', 'mentalhealth', 'community', 'movingthreat', 'domestic', 'defensive', 'h_and_w', 'offdutypistol', 'Complaint Id', 'Incident Date']]

# Save the combined data to a new CSV file
combined_data.to_csv('officer_complaints_and_trainings.csv', index=False)

# Display the first few rows of the combined data (optional)
print(combined_data.head())
```

```python
import pandas as pd

# Load the datasets
trainings = pd.read_csv('training_clean.csv')
complaints = pd.read_csv('combined_allegations_complaints.csv')

# Drop the unnecessary column
trainings = trainings.drop(columns=['timestamp..2024.07.07'])

# Convert the date columns to datetime format without specifying the format
trainings['date'] = pd.to_datetime(trainings['date'])
complaints['Incident Date'] = pd.to_datetime(complaints['Incident Date'])

# Convert both date columns to a consistent format (e.g., 'YYYY-MM-DD')
trainings['date'] = trainings['date'].dt.strftime('%Y-%m-%d')
complaints['Incident Date'] = complaints['Incident Date'].dt.strftime('%Y-%m-%d')

# Convert the dates back to datetime format after standardizing
trainings['date'] = pd.to_datetime(trainings['date'])
complaints['Incident Date'] = pd.to_datetime(complaints['Incident Date'])

# Filter trainings in May 2017
trainings_filtered = trainings[(trainings['date'] >= '2017-05-01') & (trainings['date'] < '2017-06-01')]

# Filter complaints within 6 months from May 2017
complaints_filtered = complaints[(complaints['Incident Date'] >= '2017-06-01') & (complaints['Incident Date'] < '2017-12-01')]

# Merge the datasets on Tax ID
merged_data = pd.merge(trainings_filtered, complaints_filtered, left_on='taxid', right_on='Tax ID')

# Reorder columns to have Tax ID first
merged_data = merged_data[['taxid'] + [col for col in merged_data.columns if col != 'taxid']]

# Save the result as a new CSV file
merged_data.to_csv('officer_complaints_and_trainings.csv', index=False)
```

```python
import pandas as pd

# Load the datasets
trainings = pd.read_csv('aggregated_training_data.csv')
complaints = pd.read_csv('combined_allegations_complaints.csv')

# Remove the .0 from taxid in the trainings data and convert to integer, then to string to match Tax ID format
trainings['taxid'] = trainings['taxid'].astype(int).astype(str)

# Convert the Tax ID in complaints to string to match the format
complaints['Tax ID'] = complaints['Tax ID'].astype(int).astype(str)

# Drop the date columns
trainings = trainings.drop(columns=['date'])
complaints = complaints.drop(columns=['Incident Date'])

# Calculate the total number of complaints per Tax ID
complaints_count = complaints.groupby('Tax ID').size().reset_index(name='total_complaints')

# Merge the training data with the complaints count on Tax ID
merged_data = pd.merge(trainings, complaints_count, left_on='taxid', right_on='Tax ID', how='left')

# Remove the duplicate Tax ID column from the complaints data
merged_data = merged_data.drop(columns=['Tax ID'])

# Reorder columns to have Tax ID first
merged_data = merged_data[['taxid'] + [col for col in merged_data.columns if col != 'taxid']]

# Save the result as a new CSV file
merged_data.to_csv('officer_complaints_and_trainings.csv', index=False)
```
